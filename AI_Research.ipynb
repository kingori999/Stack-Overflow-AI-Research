```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

try:
    df = pd.read_csv('survey_results_public (1).csv', usecols=['AISelect'])
except FileNotFoundError:
    print("Error: The file 'survey_results_public (1).csv' was not found.")

valid_responses = {"Yes", "No, but I plan to soon", "No, and I don't plan to"}

filtered_responses = df['AISelect'][df['AISelect'].isin(valid_responses)]

response_counts = filtered_responses.value_counts()

total_responses = len(filtered_responses)
response_percentages = (response_counts / total_responses) * 100

plt.figure(figsize=(10, 6))
response_percentages.plot(kind='bar', color='royalblue')

plt.title("Percentage of Responses for AI Tool Usage in Development Process")
plt.xlabel("Response")
plt.ylabel("Percentage of Total Valid Responses (%)")

for i, percentage in enumerate(response_percentages):
    plt.text(i, percentage + 1, f'{percentage:.1f}%', ha='center')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

try:
    df = pd.read_csv('survey_results_public (1).csv', usecols=['AISent'])
except FileNotFoundError:
    print("Error: The file 'survey_results_public (1).csv' was not found.")

valid_responses_favorability = [
    "Very favorable",
    "Favorable",
    "Indifferent",
    "Unfavorable",
    "Very unfavorable",
    "Unsure"
]

filtered_responses_favorability = df['AISent'][df['AISent'].isin(valid_responses_favorability)]

response_counts_favorability = filtered_responses_favorability.value_counts()
response_counts_favorability = response_counts_favorability.reindex(valid_responses_favorability, fill_value=0)

total_responses_favorability = len(filtered_responses_favorability)
response_percentages_favorability = (response_counts_favorability / total_responses_favorability) * 100

plt.figure(figsize=(10, 6))
bars = response_percentages_favorability.plot(kind='bar', color='royalblue')

plt.title("Favorability of Using AI Tools in Development Workflow")
plt.xlabel("Response")
plt.ylabel("Percentage of Total Valid Responses (%)")

for i, percentage in enumerate(response_percentages_favorability):
    plt.text(i, percentage + 1, f'{percentage:.1f}%', ha='center')

plt.text(len(valid_responses_favorability) - 1, max(response_percentages_favorability) + 5,
         f'Total valid responses: {total_responses_favorability}',
         ha='center', fontsize=12, color='black', weight='bold')

plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

%matplotlib inline

df = pd.read_csv('survey_results_public (1).csv', usecols=['AIBen'])

benefits = [
    "Increase productivity",
    "Greater efficiency",
    "Make workload more manageable",
    "Improve collaboration",
    "Speed up learning",
    "Improve accuracy in coding",
    "None of the above",
    "Other (please specify):"
]

valid_responses = df['AIBen'].dropna()

benefit_counts = Counter(
    benefit.strip()
    for response in valid_responses
    for benefit in response.split(',')
)

benefit_counts_ordered = {benefit: benefit_counts.get(benefit, 0) for benefit in benefits}

total_responses = len(valid_responses)
benefit_percentages = {k: (v / total_responses) * 100 for k, v in benefit_counts_ordered.items()}

benefit_percentages_series = pd.Series(benefit_percentages)

plt.figure(figsize=(12, 6))
benefit_percentages_series.plot(kind='bar', color='royalblue')

plt.title("Most Important Benefits of Using AI Tools in Development Workflow")
plt.xlabel("Benefits")
plt.ylabel("Percentage of Respondents (%)")

for i, percentage in enumerate(benefit_percentages_series):
    plt.text(i, percentage + 1, f'{percentage:.1f}%', ha='center')

plt.text(len(benefits) - 1, max(benefit_percentages_series) + 5,
         f'Total valid responses: {total_responses}',
         ha='center', fontsize=12, color='black', weight='bold')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
